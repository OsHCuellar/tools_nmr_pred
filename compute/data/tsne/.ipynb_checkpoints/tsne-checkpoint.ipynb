## tSNE creation
This script creates tSNEs and visualises them. Different labels for colouring can be chosen.

```python
import pandas as pd
import numpy as np
import os
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import rdkit
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import Descriptors
```

Some functions for the ease of the descriptor and representation generation.

```python
def mf_rdkit_des(smiles, descriptors):
    """
    Function that gets the morgan fingerprints for the smiles listed in the loaded df (creating the representation X)
    and adds easily accessible descriptors to a dictionary/ .csv file (potential labels, also returned)
    smiles : list of smile strings
    descriptors: empty dictionary
    """
    n2 = 1024  # length of the vector, first assume 1024 bit from MF
    n1 = len(smiles)
    X = np.zeros(shape=(n1, n2), dtype=np.float64)
    df = pd.DataFrame(
        columns=[
            "Smiles",
            "MolecularWeight",
            "NumHeavyAtoms",
            "NumHAcceptors",
            "NumHDonors",
            "LogP",
            "TPSA",
            "NumRotatableBonds",
            "RingCount",
            "NumAromaticRings",
        ]
    )
    for i, smi in enumerate(smiles):
        #print(i, smi)
        # make the rep
        mol = Chem.MolFromSmiles(smi)
        if not os.path.exists("descriptors.csv"):
            try:
                descriptors[i] = {
                    "Smiles": smi,
                    "MolecularWeight": Descriptors.MolWt(mol),
                    "NumHeavyAtoms": Descriptors.HeavyAtomCount(mol),
                    "NumHAcceptors": Descriptors.NumHAcceptors(mol),
                    "NumHDonors": Descriptors.NumHDonors(mol),
                    "LogP": Descriptors.MolLogP(mol),
                    "TPSA": Descriptors.TPSA(mol),
                    "NumRotatableBonds": Descriptors.NumRotatableBonds(mol),
                    "RingCount": Descriptors.RingCount(mol),
                    "NumAromaticRings": Descriptors.NumAromaticRings(mol),
                }
            except:
                descriptors[i] = {
                    "Smiles": smi,
                    "MolecularWeight": np.nan,
                    "NumHeavyAtoms": np.nan,
                    "NumHAcceptors": np.nan,
                    "NumHDonors": np.nan,
                    "LogP": np.nan,
                    "TPSA": np.nan,
                    "NumRotatableBonds": np.nan,
                    "RingCount": np.nan,
                    "NumAromaticRings": np.nan,
                }
            new_row = pd.DataFrame(descriptors[i], index=[0])
            df = pd.concat([df, new_row], ignore_index=True)
            df.to_csv("descriptors.csv")
        else:
            df = pd.read_csv("descriptors.csv")
            descriptors[i] = df.iloc[i].to_dict()
        try:
            mf = AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=1024)
            X[i, :] = mf
        except:
            empty = np.zeros(shape=(1024), dtype=np.float16)
            X[i, :] = empty
            print("appended empty")
    return X, descriptors
```

```python
def get_mf(smiles):
    """
    Function that gets the morgan fingerprints for the smiles listed in the loaded df (creating the representation X)
    smiles : list of smile strings
    """
    n2 = 1024  # length of the vector, first assume 1024 bit from MF
    n1 = len(smiles)
    X = np.zeros(shape=(n1, n2), dtype=np.float64)
    for i, smi in enumerate(smiles):
        #print(i, smi)
        # make the rep
        mol = Chem.MolFromSmiles(smi)
        try:
            mf = AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=1024)
            X[i, :] = mf
        except:
            empty = np.zeros(shape=(1024), dtype=np.float16)
            X[i, :] = empty
            print("appended empty")
    return X
```

```python
def get_target(descriptors, descriptorname, y):
    """
    Gets the label for the tsne if one of the descriptors is chosen and returns it in the np.array y
    descriptors: dictionary from the previous funciton
    descriptorname: which descriptor to use, string
    y: empty list
    """
    for i, elem in enumerate(list(descriptors.keys())):
        y.append(descriptors[elem][descriptorname])
    y = np.array(y)
    return y
```

```python
# This cell is for using the latent space created from the GNN (DO NOT RUN IF MF ARE CHOSEN)/ TURN INTO CODE IF CHOSEN
########################################## 
# HERE INPUT 
# decide which latent space (mw_low, mw_high, random)
# what to call the representation
# and what to use as target/ label
dir_to_use = "mw_low"
rep = "latent_space"
target = "peakcounter"
##########################################

names = np.load(f"../results/{dir_to_use}/smiles.npy")
# Find the matching indices
matching_indices = np.where(np.isin(smiles, names))[0]


X_MF, descriptors = mf_rdkit_des(smiles, descriptors)

X = np.load(f"../results/{dir_to_use}/latent.npy")

# load the rest
y_predict = np.load(f"../results/{dir_to_use}/prediction.npy")
y_target = np.load(f"../results/{dir_to_use}/target.npy")

MAE = [abs(x - y) for x, y in zip(y_predict, y_target)]

########################################## 
# HERE INPUT 
# decide which label to actually use (y_predict, y_target, MAE)
#y_preselection = get_target(descriptors, target, y)
#Select the corresponding values from y_preselection IF we use the the X=latent space
#y = y_preselection[matching_indices]

#for y=MAE,y_predict or y_target and X=latent space, no selection is needed
y = y_target

##########################################
```

Now define the distance metrices (how to measure similarity) and perplexities (how many points are expected to be within one cluster/ amount of nearest neighbours). The colour limits for the colourbar should also be adjusted according to which values one expects.

```python
########################################## 
# HERE INPUT
# Define the distance metrics and perplexities to use
metrics = ["euclidean", "manhattan", "cosine"]
perplexities = [10, 20, 40, 80, 160, 320]

# Define the colour limits for the colourbar
vmin = 0
vmax = 30
##########################################
```

The overview is now created and then plotted.

```python
# Create subplots for each combination of metric and perplexity
fig, axes = plt.subplots(nrows=len(metrics), ncols=len(perplexities), figsize=(20, 15))

# Perform t-SNE dimensionality reduction for each combination of metric and perplexity
for i, metric in enumerate(metrics):
    for j, perplexity in enumerate(perplexities):
        tsne = TSNE(
            n_components=2, random_state=0, metric=metric, perplexity=perplexity
        )
        X_tsne = tsne.fit_transform(X)

        # Plot the t-SNE map in the corresponding subplot
        scatter = axes[i, j].scatter(
            X_tsne[:, 0],
            X_tsne[:, 1],
            c=y,
            cmap="viridis",
            vmin=vmin,
            vmax=vmax,
        )  
        axes[i, j].set_title(f"{metric.capitalize()}, perplexity={perplexity}")
        axes[i, j].set_xticks([])
        axes[i, j].set_yticks([])
        fig.colorbar(scatter, ax=axes[i, j])

# Set a global title
fig.suptitle("t-SNE maps for different distance metrics and perplexities")

# Adjust the spacing between subplots
fig.tight_layout()
```

```python
# Show figure and save it
fig.savefig(f"{dir_to_use}/tsne_overview_{rep}_{target}.png")
```

After seeing the overview, we can select whatever we deem the most useful and plot it individually, which is done here.

```python
# Now do it for one and save it (the ones before are nice for a comparison, but let's keep it simple
########################################## 
# HERE INPUT
# decide based on the overview picture from before
metric = "cosine"
perp = 10
########################################## 
```

```python
# Perform t-SNE dimensionality reduction
tsne = TSNE(
    n_components=2,
    random_state=0,
    metric=metric,
    perplexity=perp,
    learning_rate=200.0,
    n_iter=1000,
)
X_tsne = tsne.fit_transform(X)
```

Saving the information for a given tsne-setting in a csv allows for facile analysis of the clusters - either with https://openmolecules.org/datawarrior/index.html or the online tool available at http://dhcp-173-dist-b-207.epfl.ch/ (from an EPFL network).

```python
# Save it in a csv for easy access and different plotting possibilities 
tsne_df = pd.DataFrame(X_tsne, columns=['x', 'y'])
tsne_df['Smiles'] = names
tsne_df[target] = y
tsne_df.to_csv(f"{dir_to_use}/tsne_{rep}_{metric}_{perp}_{target}.csv")
```

```python
# Visualize the t-SNE map
plt.figure(figsize=(10, 8))
scatter = plt.scatter(
    X_tsne[:, 0], X_tsne[:, 1], c=y, cmap="viridis", vmin=vmin, vmax=vmax
)
plt.axis("off")
plt.colorbar(scatter)
plt.savefig(f"{dir_to_use}/tsne_{rep}_{metric}_{perp}_{target}.png")
```
